```python
rain = [45, 52, 120, 156, 180, 165, 140, 135, 98, 65, 30, 28]

total = sum(rain)
avg = total / len(rain)

print("Total Rainfall:", total)
print("Average Rainfall:", avg)
print("Maximum:", max(rain))
print("Minimum:", min(rain))

print("Months above average:")
for r in rain:
    if r > avg:
        print(r)
```
