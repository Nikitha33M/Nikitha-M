```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Sample data for demonstration. In a real scenario, you would load your actual data here.
data = {'Rainfall_mm': [100, 150, 200, 250, 300],
        'Yield_kg': [500, 650, 800, 950, 1100]}
df = pd.DataFrame(data)

X = df[["Rainfall_mm"]]
y = df["Yield_kg"]
# Very simple model
model_simple = LinearRegression()
model_simple.fit(X, y)
model_simple.predict([[205]])
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
# Very complex model
model_complex = make_pipeline(
 PolynomialFeatures(degree=5),
 LinearRegression()
)
model_complex.fit(X, y)
model_complex.predict([[205]])
```
